```python
df1 = df.head(50)
df1
```

```python
default_x_ticks = range(len(df1.stackoverflow_id))
```

```python
plt.figure(figsize=(14,6))
plt.plot(default_x_ticks, df1.question_count/100,)
plt.xticks(default_x_ticks, df1.stackoverflow_id,rotation=90)
plt.plot(default_x_ticks, df1.reputation/10**5)
plt.xticks(default_x_ticks, df1.stackoverflow_id,rotation=90)
plt.grid()
plt.legend(['Questions asking','Reputation'])
plt.xlabel('Users')
plt.ylabel('Questions asking')
```

```python
plt.figure(figsize=(14,6))
plt.plot(default_x_ticks, df1.answer_count/100,)
plt.xticks(default_x_ticks, df1.stackoverflow_id,rotation=90)
plt.plot(default_x_ticks, df1.reputation/10**5)
plt.xticks(default_x_ticks, df1.stackoverflow_id,rotation=90)
plt.grid()
plt.legend(['Answers','Reputation'])
plt.xlabel('Users')
plt.ylabel('Question replying')
```

```python
plt.figure(figsize=(14,6))
plt.plot(default_x_ticks, df1.question_count/1000,)
plt.xticks(default_x_ticks, df1.stackoverflow_id,rotation=90)
plt.plot(default_x_ticks, df1.Openness)
plt.xticks(default_x_ticks, df1.stackoverflow_id,rotation=90)
plt.grid()
plt.legend(['Question asking','Openness'])
plt.xlabel('Users')
plt.ylabel('Scores')
```
